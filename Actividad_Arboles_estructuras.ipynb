```python
import random # Importa el módulo random para generar datos aleatorios.
import json   # Importa el módulo json para guardar datos en formato JSON.

# Lista de nombres posibles para los estudiantes.
nombres = [
    "Ana", "Carlos", "Maria", "Luis", "Sofia", "Pedro", "Laura",
    "Andres", "Valentina", "Diego", "Camila", "Mateo", "Daniela", "Esteban", "Carlos"
]

estudiantes = [] # Inicializa una lista vacía para almacenar los diccionarios de estudiantes.

# Bucle para generar 10,000 estudiantes.
for i in range(10000):

    # Crea un diccionario para cada estudiante con id, nombre y promedio aleatorios.
    estudiante = {
        "id": random.randint(1000, 9999),          # Genera un ID aleatorio entre 1000 y 9999.
        "nombre": random.choice(nombres),          # Selecciona un nombre aleatorio de la lista 'nombres'.
        "promedio": round(random.uniform(0, 10),2) # Genera un promedio aleatorio entre 0 y 10, redondeado a 2 decimales.
    }

    estudiantes.append(estudiante) # Añade el diccionario del estudiante a la lista 'estudiantes'.

# Guarda la lista de estudiantes en un archivo JSON.
with open("estudiantes.json", "w") as archivo:
    json.dump(estudiantes, archivo, indent=4) # Serializa la lista a JSON y la escribe en el archivo con indentación.

print("Archivo generado con 10000 estudiantes.") # Muestra un mensaje de confirmación.
```

```python
def lista_desordenada(estudiantes):
    # Esta función simplemente retorna la lista de estudiantes tal como está (desordenada).
    return estudiantes

desordenados = lista_desordenada(estudiantes) # Asigna la lista original de estudiantes a 'desordenados'.

print(desordenados[:100]) # Imprime los primeros 100 estudiantes de la lista desordenada para visualización.
```

```python
def ordenar_estudiantes(estudiantes):
    # Ordena la lista de estudiantes basándose en el valor de la clave 'id' de cada diccionario.
    # Utiliza una función lambda como 'key' para especificar el criterio de ordenación.
    return sorted(estudiantes, key=lambda x: x["id"])
```

```python
class Nodo:
    def __init__(self, estudiante):
        self.estudiante = estudiante # Almacena el diccionario del estudiante en el nodo.
        self.izquierda = None        # Puntero al nodo hijo izquierdo.
        self.derecha = None          # Puntero al nodo hijo derecho.


class ArbolABB:

    def __init__(self):
        self.raiz = None # Inicializa la raíz del árbol como vacía.

    # Método público para insertar un estudiante.
    def insertar(self, estudiante):
        if self.raiz is None:
            self.raiz = Nodo(estudiante) # Si el árbol está vacío, el nuevo estudiante se convierte en la raíz.
        else:
            self._insertar(self.raiz, estudiante) # Llama al método auxiliar para la inserción recursiva.

    # Método auxiliar recursivo para insertar un estudiante.
    def _insertar(self, nodo, estudiante):
        # Compara el ID del nuevo estudiante con el ID del estudiante en el nodo actual.
        if estudiante["id"] < nodo.estudiante["id"]:
            if nodo.izquierda is None:
                nodo.izquierda = Nodo(estudiante) # Si el ID es menor y no hay hijo izquierdo, inserta aquí.
            else:
                self._insertar(nodo.izquierda, estudiante) # Si hay hijo izquierdo, continúa la búsqueda recursivamente por la izquierda.
        else:
            if nodo.derecha is None:
                nodo.derecha = Nodo(estudiante)   # Si el ID es mayor o igual y no hay hijo derecho, inserta aquí.
            else:
                self._insertar(nodo.derecha, estudiante) # Si hay hijo derecho, continúa la búsqueda recursivamente por la derecha.

    # Método público para buscar un estudiante por ID.
    def buscar(self, id):
        return self._buscar(self.raiz, id) # Llama al método auxiliar para la búsqueda recursiva.

    # Método auxiliar recursivo para buscar un estudiante.
    def _buscar(self, nodo, id):
        if nodo is None:
            return None # Si el nodo es nulo, el estudiante no se encuentra.

        if nodo.estudiante["id"] == id:
            return nodo.estudiante # Si el ID coincide, retorna el estudiante.

        elif id < nodo.estudiante["id"]:
            return self._buscar(nodo.izquierda, id) # Si el ID es menor, busca en el subárbol izquierdo.

        else:
            return self._buscar(nodo.derecha, id) # Si el ID es mayor, busca en el subárbol derecho.

    # Método público para listar los estudiantes en orden (recorrido inorden).
    def listar(self):
        self._inorden(self.raiz) # Llama al método auxiliar para el recorrido inorden.

    # Método auxiliar recursivo para el recorrido inorden.
    def _inorden(self, nodo):
        if nodo:
            self._inorden(nodo.izquierda) # Recorre el subárbol izquierdo.
            print(nodo.estudiante)        # Visita el nodo actual (imprime el estudiante).
            self._inorden(nodo.derecha)  # Recorre el subárbol derecho.

    # Método para insertar una lista completa de estudiantes.
    def insertar_lista(self, lista_estudiantes):
        for estudiante in lista_estudiantes:
            self.insertar(estudiante) # Itera sobre la lista e inserta cada estudiante individualmente.
```

```python
class NodoBPlus:

    def __init__(self, hoja=False):
        self.hoja = hoja     # Indica si el nodo es una hoja (contiene datos) o un nodo interno.
        self.claves = []     # Lista de claves (en un nodo hoja, son los estudiantes; en un nodo interno, son los IDs de separación).
        self.hijos = []      # Lista de punteros a nodos hijos (solo relevante para nodos internos).


class ArbolBPlus:

    def __init__(self):
        self.raiz = NodoBPlus(True) # Inicializa la raíz como un nodo hoja (versión simplificada).

    # Método para buscar un estudiante por ID en el árbol B+ (simplificado).
    def buscar(self, id):

        nodo = self.raiz # Comienza la búsqueda desde la raíz.

        # En esta implementación simplificada, solo hay un nivel (la raíz es una hoja).
        # Un B+ real tendría un bucle 'while not nodo.hoja' para descender por los niveles.

        # Itera sobre las claves (estudiantes) en el nodo hoja.
        for estudiante in nodo.claves:
            if estudiante["id"] == id:
                return estudiante # Si encuentra el ID, retorna el estudiante.

        return None # Si el estudiante no se encuentra después de recorrer las claves.

    # Método para insertar un estudiante (versión simplificada).
    def insertar(self, estudiante):

        nodo = self.raiz # Obtiene el nodo raíz (que es una hoja en esta implementación).

        nodo.claves.append(estudiante) # Añade el nuevo estudiante a la lista de claves del nodo raíz.

        nodo.claves.sort(key=lambda x: x["id"]) # Mantiene las claves ordenadas por ID después de cada inserción.

    # Método para listar todos los estudiantes en el árbol B+ (simplificado).
    def listar(self):

        for e in self.raiz.claves: # Itera y imprime cada estudiante en la lista de claves de la raíz.
            print(e)

    # Método para insertar una lista completa de estudiantes.
    def insertar_lista(self, lista_estudiantes):
        for estudiante in lista_estudiantes:
            self.insertar(estudiante) # Itera sobre la lista e inserta cada estudiante individualmente.
```

```python
arbolBplus = ArbolBPlus() # Crea una instancia del Árbol B+.

arbolBplus.insertar_lista(desordenados) # Inserta todos los estudiantes de la lista 'desordenados' en el ArbolBplus.

print("ArbolBPlus poblado con éxito.") # Muestra un mensaje de confirmación.

print(arbolBplus.buscar(1001)) # Busca y muestra el estudiante con ID 1001 en el ArbolBplus.

arbolBplus.listar() # Lista todos los estudiantes en el ArbolBplus en orden (por ID).
```

```python
def buscar_estudiante(estudiantes, id_buscar):
    # Itera sobre cada estudiante en la lista.
    for e in estudiantes:
        if e["id"] == id_buscar: # Si el ID del estudiante coincide con el ID buscado...
            return e              # ...retorna el diccionario del estudiante.
    return None                   # Si ningún estudiante coincide después de revisar toda la lista, retorna None.
```

```python
import time # Importa el módulo time para medir tiempos de ejecución.

def comparar_estructuras_single_id(desordenados, ordenados, arbolABB, arbolBplus, id_a_buscar):

    resultados = {} # Diccionario para almacenar los tiempos de cada estructura.

    # Búsqueda en lista desordenada
    inicio = time.time() # Marca el inicio del tiempo.
    buscar_estudiante(desordenados, id_a_buscar) # Ejecuta la búsqueda.
    resultados["lista_desordenada"] = time.time() - inicio # Calcula el tiempo transcurrido y lo guarda.

    # Búsqueda en lista ordenada
    inicio = time.time()
    buscar_estudiante(ordenados, id_a_buscar) # Ejecuta la búsqueda en la lista ordenada.
    resultados["lista_ordenada"] = time.time() - inicio

    # Búsqueda en ABB
    inicio = time.time()
    arbolABB.buscar(id_a_buscar) # Ejecuta la búsqueda en el ArbolABB.
    resultados["ABB"] = time.time() - inicio

    # Búsqueda en B+
    inicio = time.time()
    arbolBplus.buscar(id_a_buscar) # Ejecuta la búsqueda en el ArbolBplus.
    resultados["B+"] = time.time() - inicio

    return resultados # Retorna el diccionario con los tiempos de cada estructura.
```
